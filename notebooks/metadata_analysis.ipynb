```python
import pandas as pd
import numpy as np
from collections import Counter
import json
import matplotlib.pyplot as plt
import sys
import os
```

```python
# Check if file exists
with open("books_metadata.json", "r") as f:
    data = json.load(f)

df = pd.DataFrame(data)

print("First rows of dataset:")
print(df.head())

# Number of unique authors
num_authors = df['author'].nunique()
print(f"\nNumber of different authors: {num_authors}")

# Years distribution
year_counts = df['year'].value_counts().sort_index()
print("\nYear distribution:")
print(year_counts)

# Plot histogram of years
plt.figure(figsize=(10,4))
plt.hist(df['year'], bins=range(int(df['year'].min()), int(df['year'].max())+2), edgecolor='black')
plt.title("Publication Years Distribution")
plt.xlabel("Year")
plt.ylabel("Count")
plt.savefig("books_plot.png", dpi=600, bbox_inches="tight")
plt.show()

# Year range
year_range = (df['year'].min(), df['year'].max())
print(f"\nRange of years: {year_range[0]} - {year_range[1]}")
```
